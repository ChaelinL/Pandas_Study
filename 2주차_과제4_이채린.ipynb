### **49. null 값이 있는 행 삭제하기**


#### **1) 모두 null 값으로 이뤄진 열 삭제하기**

**'Unnamed: 19' 열은 전부 NaN값으로 이루어져있음을 확인할 수 있었습니다. 이 열은 불필요할 것 같으니 제거해줍시다.행을 삭제할 것인지 열을 삭제할 것인지는 메서드의 axis 옵션을 통해 설정 가능합니다.** 

**우리는 결측치 비율을 구하는 과정을 통해 해당 열이 전부 NaN값으로 이루어져있다는 사실을 이미 알았지만, 메서드의 how 옵션을 통해 전부 결측치로 이루어진 열만 삭제할 수 있습니다다.**

```python
df = df.dropna(axis=1, how='all')
df
```

**19 columns라고 값이 바뀌었네요. 'Unnamed: 19' 열이 잘 지워진 것을 확인할 수 있습니다.**

#### **2) 결측치가 특정 개수, 특정 비율 이상안 열/행 drop하기**

**'소재지도로명주소' 열 또한 55%나 결측치로 이루어진 것을 확인할 수 있었습니다. 특정 열이나 행에서 결측치의 값이 특정 개수 이상이거나, 특정 비율 이상이라면 drop하도록 할 수도 있습니다.**

```python
# 결측치의 비율이 50% 이상인 열을 제거해봅시다. 
# 위에서 구한 결측치 비율을 활용하여 제거할 열을 뽑고 drop시키면 됩니다.

df2 = df.drop(columns=df.columns[pd.isnull(df).mean()*100 >= 0.5], axis=1)

# 얼마나 많은 열이 제거되었는지 확인해봅시다.
print(len(df.columns),len(df2.columns))
```

```python
# 결측치의 개수가 7개 이상인 행을 제거해봅시다. 메서드의 thresh 옵션을 통해 설정할 수 있습니다.
df3 = df.dropna(thresh=13, axis=0)  # thresh => 축을 기준으로 n개 미만 입력되면 해당 축 제거 (따라서 20개의 행 중 결측치가 7개 이상이려면 n=13)

# 얼마나 많은 행이 제거되었는지 확인해봅시다.
print(df.shape,df3.shape)
```

### **50. null 값을 특정 값으로 대체하기**
위에서 열 별로 null 값의 수와 비율을 계산하면서, 도로명 주소의 결측치 비율이 55%나 되는 것을 확인하였습니다. 이를 대체하기 위해서 도로명 주소와 지번 주소 중 하나만 있을 때는 나머지 데이터로 결측치를 대체해봅시다.

```python
# 도로명 주소의 널값 수
df['소재지도로명주소'].isnull().sum()
```

```python
# 지번 주소의 널값 수
df['소재지지번주소'].isnull().sum()
```

```python
# "소재지도로명주소"와 "소재지지번주소"가 모두 결측치가 아닌 데이터를 찾습니다.
df[(df['소재지도로명주소'].notnull() & df['소재지지번주소'].notnull())].shape
```

```python
# "소재지도로명주소"의 결측치를 fillna 를 통해 "소재지지번주소"로 채웁니다.
df['소재지도로명주소'] = df['소재지도로명주소'].fillna('소재지지번주소')
df
```

```python
# "소재지도로명주소"의 결측치수를 세어봅니다.
df['소재지도로명주소'].isnull().sum()
```

## **2. 결측치를 처리하는 다른 여러 방법**
결측치를 임의의 동일한 값으로 채우지 않고 평균이나 표준편차 값을 계산하여 대체하는 것을 교재에서 살펴보았습니다. 하지만 결측치를 앞의 값이나 뒤의 값으로 채우거나, 보간(interpolation)법을 사용한 계산 결과로도 채울 수 있습니다.

### **51. null 값을 특정 계산 결과로 대체하기** 

**1) 보간법을 사용하여 채우기**

시계열 데이터나 연속된 수치를 가지는 데이터의 경우에는 일종의 **연속성있는 패턴**을 발견할 수 있습니다. 이런 경우 보간(Interpolation)을 통해 앞, 뒤 값을 가지고 유추하여 좀 더 잘 결측치를 채워줄 수 있습니다.  

보간은 DataFrame이나 Series에 interpolate 메서드를 적용하면 바로 결측치가 채워집니다. method 옵션을 사용하여서 어떤 값을 기준으로 할 지 지정할 수 있습니다.

```python
from pandas import DataFrame, Series
from datetime import datetime
datestrs = ['12/1/2016', '12/03/2016', '12/04/2016', '12/10/2016']
dates = pd.to_datetime(datestrs)
ts = Series([1, np.nan, np.nan, 10], index=dates)
```

```python
# 시계열데이터의 값에 선형으로 비례하는 방식으로 결측값을 보간합니다.
ts_intp_linear = ts.interpolate()
ts_intp_linear
```

```python
# 시계열 날짜 인덱스를 기준으로 결측값을 보간합니다.
ts_intp_time = ts.interpolate(method='time')
ts_intp_time
```

```python
# NaN 값이 포함되어 있는 새로운 데이터프레임을 만듭니다.
df4 = DataFrame({'C1': [1, 2, np.nan, np.nan, 5],
     'C2': [6, 8, 10, np.nan, 20]})
df4
```

```python
# 데이터프레임 값에 선형으로 비례하는 방식으로 결측치를 보간합니다.
df_intp_values = df4.interpolate()
df_intp_values
```

**2) 앞의 값이나 뒤의 값으로 채우기**

fillna 메서드의 method 옵션을 조정하면 앞의 값이나 뒤의 값으로 결측치를 채울 수 있습니다.

```python
# 앞의 값으로 채워봅시다.
df4.fillna(method='ffill')
```

```python
df4 = DataFrame({'C1': [1, 2, np.nan, np.nan, 5],
     'C2': [6, 8, 10, np.nan, 20]})
```

```python
# 뒤의 값으로  채워봅시다.
df4.fillna(method='bfill')
```
