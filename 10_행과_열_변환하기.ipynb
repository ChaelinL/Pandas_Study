```python
import pandas as pd

df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
df
```

**모든 열을 행으로 변환하기**

```python
pd.melt(df)
```

**하나의 열만 행으로 이동시키기**

```python
pd.melt(df, id_vars=['A'], value_vars=['B'])  # A열만 그대로, B열은 행으로 이동
```

**여러 개의 열을 행으로 이동시키기**

```python
df_melt= pd.melt (df, id_vars=['A'], value_vars=['B','C'])  # A열만 그대로, B열과 C열은 행으로 이동
df_melt
```

**특정 열의 값을 기준으로 새로운 열 만들기**

```python
df_pivot = df_melt.pivot(index='A', columns='variable', values='value')  # A열은 새로운 인덱스로 만들고, B열과 C열은 이에 따라 정리
df_pivot
```

**원래 데이터 형태로 되돌리기**

```python
df_origin = df_pivot.reset_index()  # 인덱스를 리셋함
df_origin.columns.name = None  # 인덱스 열의 이름을 초기화함
df_origin
```
