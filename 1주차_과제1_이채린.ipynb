##4.데이터의 열 이름을 따로 지정해서 만들기
2번 문제의 데이터 프레임을 열 이름을 따로 지정하여 만들어보자.

image.png

```python
df = pd.DataFrame(
    [[0,3,'male',22.0],
     [1,1,'female',38.0],
     [1,3,'female',26.0],
     [1,1,'female',35.0],
     [0,3,'male',35.0]],
     columns=['survived', 'pclass', 'sex', 'age']
)

df
```

##5.인덱스가 두 개인 데이터 프레임 만들기
인덱스를 튜플로 지정하고, 인덱스 이름을 지정하여 다음 사진과 같은 데이터 프레임을 만들어보자.

image.png

```python
df = pd.DataFrame(
    {"survived":[0,1,1,1,0],
     "pclass":[3,1,3,1,3],
     "sex":['male','female','female','female','male'],
     "age":[22.0,38.0,26.0,35.0,35.0]},
     index=pd.MultiIndex.from_tuples(
         [('a',1),('b',2),('b',3),('c',4),('c',5)],
         names=['n', 'PassengerId']
     )
)
```
