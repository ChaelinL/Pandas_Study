```python
import pandas as pd

df = pd.DataFrame(
    {"a": [4, 5, 6, 7],
     "b": [8, 9, 10, 11],
     "c": [12, 13, 14, 15]},
     index=pd.MultiIndex.from_tuples(
         [('d',1), ('d', 2), ('e', 1), ('e', 2)],
         names=['n', 'v']
     )
)

df
```

**인덱스의 범위로 불러오기**

```python
df.iloc[1:4]  # 두번째 인덱스부터 순서대로 3개를 불러옴
```

**첫 인덱스를 지정해 불러오기**

```python
df.iloc[2:]  # 세번째 인덱스부터 마지막 인덱스까지 불러옴
```

**마지막 인덱스를 지정해 불러오기**

```python
df.iloc[:3]  # 처음부터 세번째 인덱스까지 불러옴
```

**모든 인덱스 불러오기**

```python
df.iloc[:]
```

**특정 열을 지정해 가져오기**

```python
df[['a','b']]  # a열과 b열을 가져옴
```

**조건을 만족하는 열 가져오기**

```python
df.filter(regex='c')  # 열 이름에 c라는 문자가 포함되어 있으면 출력함
```

**특정 문자가 포함되지 않는 열 가져오기**

```python
df.filter(regex='^(?!c$).*')  # 열 이름에 c라는 문자가 포함되어 있지 않으면 출력함
```
