```python
import pandas as pd

df = pd.DataFrame(
    {"a": [4, 5, 6, 7],
     "b": [8, 9, 10, 11],
     "c": [12, 13, 14, 15]},
     index=pd.MultiIndex.from_tuples(
         [('d',1), ('d', 2), ('e', 1), ('e', 2)],
         names=['n', 'v']
     )
)
df
```

**조건에 맞는 새 열 만들기**

```python
df['sum'] = df['a']+df['b']+df['c']
df
```

**assign()을 이용해 조건에 맞는 새 열 만들기**

```python
# a열, b열, c열의 값을 모두 더해 새로운 열 만들기
df = df.assign(multiply=lambda df: df['a']*df['b']*df['c'])
df
```

**숫자형 데이터를 구간으로 나누기**

```python
# a열을 두 개로 나누어 각각 새롭게 레이블 만들기
df['qcut'] = pd.qcut(df['a'], 2, labels=["600이하", "600이상"])
df
```

**기준값 이하와 이상을 모두 통일시키기**

```python
# a열에서 5 이하는 모두 5로, 6 이상은 모두 6으로 변환
df['clip'] = df['a'].clip(lower=5, upper=6)
df
```

**최댓값 불러오기**

```python
df.max(axis=0)  # axis=0은 행(a, b, c) 비교, axis=1은 열(d, e) 비교 
```

**최솟값 불러오기**

```python
df.min(axis=0)
```
