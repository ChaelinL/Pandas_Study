```python
import pandas as pd

df = pd.DataFrame(
    {"a": [4, 5, 6, 7],
     "b": [8, 9, 10, 11],
     "c": [12, 13, 14, 15]},
     index=pd.MultiIndex.from_tuples(
         [('d',1), ('d', 2), ('e', 1), ('e', 2)],
         names=['n', 'v']
     )
)
```

**특정 행과 열을 지정해 가져오기**

```python
# df.loc[가져올 행, 가져올 열]
df.loc[:,'a':'c']  # 모든 인덱스에서 a열부터 c열까지 가져오라는 뜻
```

**인덱스로 특정 행과 열 가져오기**

```python
df.iloc[0:3,[0,2]]  # 0 인덱스부터 2 인덱스까지 0번째 열과 2번째 열을 가져오라는 의미 (첫 열이 0번째)
```

**특정 열에서 조건을 만족하는 행과 열 가져오기**

```python
df.loc[df['a']>5, ['a','c']]  # a열 값이 5보다 큰 경우 a열과 c열을 출력하라는 뜻
```

**인덱스를 이용해 특정 조건을 만족하는 값 불러오기**

```python
df.iat[1,2]  # 1번째 인덱스에서 2번째 열 값을 가져옴
```
