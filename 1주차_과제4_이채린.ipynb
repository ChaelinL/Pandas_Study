#**데이터 분석을 위한 판다스 과제**

다음은 붓꽃(iris)의 3가지 종(setosa, versicolor, virginica)에 대한 꽃받침(sepal)과 꽃잎(petal)의 길이(length)와 너비(width)를 정리한 데이터입니다.  
주어진 코드를 이용하여 문제를 풀어 보세요.
<br/><br/>
\* ___ 안에 들어갈 코드를 작성하시면 됩니다._

```python
import pandas as pd
import seaborn as sns
df = sns.load_dataset('iris')   # seaborn 패키지에서 사용할 수 있는 내장 데이터 중 iris 데이터셋(pandas dataframe 형태)을 불러옵니다.
df.head()
```

####**1. 처음부터 5번째 인덱스까지 불러오시오.**

```python
df.iloc[:5]
```

####**2. 모든 인덱스를 불러오시오.**

```python
df.iloc[:]
```

####**3. 종(species) 열의 데이터만 가져오시오.**

```python
df.iloc[:,-1:]    # 행은 전체 추출하고, 열은 마지막 열만 추출합니다.
```

####**4. 꽃잎(petal)에 대한 정보를 담고 있는 열을 가져오시오.**

```python
df[['petal_length','petal_width']]
```

####**5. 꽃받침(sepal)에 대한 정보를 담고 있는 열을 가져오시오.**

```python
df.filter(like='sepal')    # 열 이름에 sepal 문자열을 포함하면 출력하라는 의미입니다.
```
