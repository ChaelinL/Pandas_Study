```python
import pandas as pd
from vega_datasets import data
```

```python
#데이터 불러오기
airports = data.airports()
```

```python
#데이터 잘 불러왔는지 확인(데이터를 위에서 부터 다섯줄 출력)
airports.head()
```

```python
#latitude, longitude 부분만 남기기
airports = airports.loc[:,['latitude', 'longitude']]
airports.head()
```

```python
#각 열의 합 구하기
airports.sum()
```

```python
#각 열의 값 개수 세기
airports.count()
```

```python
#데이터 중앙값 출력
airports.median()
```

```python
#데이터의 50% 지점 값 출력
airports.quantile([0.5])
```

```python
#latitude 열의 평균과 표준편차 출력
print("평균:", airports['latitude'].mean(), "\n표준편차:", airports['latitude'].std())
```

```python
import numpy as np
#데이터를 소수점 둘째자리에서 반올림. Hint! lambda 식 활용
airports.apply(lambda x: round(x,2))
```
