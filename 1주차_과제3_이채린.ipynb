##가격이 가장 비싼 10개의 보석을 출력해보자

```python
data.nlargest(10,'price')
```

##가격이 가장 싼 10개의 보석을 출력해보자

```python
data.nsmallest(10,'price')
```

##E color를 가진 가장 비싼 보석 5개를 출력해보자

```python
data[data['color']=='E'].nlargest(5,'price')
```

##x가 가장 큰 10개 보석들 중 y값이 가장 작은 값 5개를 출력해보자

```python
data.nlargest(10,'x').nsmallest(5,'y')
```

##z값이 가장 큰 보석들 10개를 가격이 높은순으로 정렬해서 출력해보자

```python
data.nlargest(10,'z').sort_values('price',ascending=False)
```

##20개의 데이터를 임의로 뽑아보자

```python
samples = data.sample(n=20)
samples
```

##뽑은 데이터의 index를 초기화시켜보자(Tip. drop=True)

```python
samples.reset_index(drop=True)
```

##뽑은 sample들을 x, y, z 순으로 정렬해보자
x가 같은 경우 y를 key로, y가 같은 경우 z를 key로 사용

```python
samples.sort_values(by=['x','y','z'], ascending=[True, True, True])
```

##다시 30개의 sample을 뽑아보자 이번에는 price에 가중치를 두어 sample을 추출한다(Tip. weights)

```python
samples2 = data.sample(n=30, weights='price')
samples2
```
