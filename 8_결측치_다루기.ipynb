```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {"a": [4, 5, 6, np.nan],
     "b": [7, 8, np.nan, 9],
     "c": [10, np.nan, 11, 12]},
     index=pd.MultiIndex.from_tuples(
         [('d',1), ('d', 2), ('e', 1), ('e', 2)],
         names=['n', 'v']
     )
)
df
```

**null 값인지 확인하기**

```python
pd.isnull(df)
```

**null 값이 아닌지 확인하기**

```python
pd.notnull(df)
```

**null 값이 있는 행 삭제하기**

```python
df.dropna()
```

**null 값을 특정 값으로 대체하기**

```python
df.fillna(13)  # null 값을 13으로 대체
```

**null 값을 특정 계산 결과로 대체하기**

```python
df.fillna(df['a'].mean())  # null 값을 a열의 평균값으로 대체
```
