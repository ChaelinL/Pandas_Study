```python
import pandas as pd
```

```python
#데이터 프레임 생성
index_tuples = [('1', '1'), ('1', '2'), ('2', '3'), ('2', '4'), ('2', '5'), ('3', '6'),('3', '7')]
values = [ [60,97,60], [30,98,58], [50,89,98], [90,78,25], [48,87,12], [65,68,98], [89,68,87]]
index = pd.MultiIndex.from_tuples(index_tuples,names=['nclass', 'id']) # 인덱스 설정
df = pd.DataFrame(values, columns=['math', 'english', 'science'], index = index)

df
```

## 데이터를 'math' 값을 기준으로 오름차순으로 정렬해보자.

```python
df.sort_values('math', ascending=True)
```

## 데이터의 'science' 열 이름을 'korean'으로 변경해보자.

```python
df.rename(columns={"science":"korean"})
```

## 데이터의 인덱스 값을 초기화해보자.

```python
df.reset_index()
```

## 데이터를 다시 인덱스 순서대로 정렬해보자.

```python
df.sort_index()
```

## 데이터의 'math', 'science' 열을 제거해보자.

```python
df.drop(columns=['math','science'])
```
