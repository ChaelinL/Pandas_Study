```python
import pandas as pd

df = pd.DataFrame(
    {"a": [4, 5, 6, 7],
     "b": [8, 9, 10, 11],
     "c": [12, 13, 14, 15]},
     index=pd.MultiIndex.from_tuples(
         [('d',1), ('d', 2), ('e', 1), ('e', 2)],
         names=['n', 'v']
     )
)
```

**특정 열 값을 기준으로 정렬하기**

```python
df.sort_values('a', ascending=False)  # ascending True는 오름차순, False는 내림차순
```

**열 이름 변경하기**

```python
df.rename(columns={"c":"d"})  # c열의 이름을 d로 변경
```

**인덱스 값 초기화하기**

```python
df.reset_index()
```

**인덱스 순서대로 정렬하기**

```python
df.sort_index()
```

**특정 열 제거하기**

```python
df.drop(columns=["a", "b"])
```
