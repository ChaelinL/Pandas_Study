```python
import pandas as pd

df = pd.DataFrame(
    {"a": [4, 5, 6, 7],
     "b": [8, 9, 10, 11],
     "c": [12, 13, 14, 15]},
     index=pd.MultiIndex.from_tuples(
         [('d',1), ('d', 2), ('e', 1), ('e', 2)],
         names=['n', 'v']
     )
)

df
```

**맨 위의 행 출력하기**

```python
df.head(2)  # 맨 위 2행만 출력
```

**맨 아래 행 출력하기**

```python
df.tail(2)  # 맨 아래 2행만 출력
```

**특정 열의 값을 추출하기**

```python
df[df["a"]>4]  # a열 중 4보다 큰 값이 있을 경우 해당 행을 추출
```

**특정 열에 특정 값이 있을 경우 추출하기**

```python
df[df["a"]==6]  # a열 중 6이 있을 경우 해당 행을 추출
```

**특정 열에 특정 값이 없을 경우 추출하기**

```python
df[df["a"] != 5]  # a열 중 5가 없을 경우 해당 행을 추출
```

**특정 열에 특정 숫자가 있는지 확인하기**

```python
df[df["a"].isin([4])]
```

**특정 비율로 데이터 샘플링하기**

```python
df.sample(frac=0.8)  # 실행할 때마다 정해진 비율만큼 랜덤하게 추출
```

**특정 개수만큼 데이터 샘플링하기**

```python
df.sample(n=3)  # 실행할 때마다 n에서 정한 만큼 랜덤하게 추출
```

**특정 열에서 큰 순서대로 불러오기**

```python
df.nlargest(3, 'a')  # a열에서 큰 순서대로 3개를 불러옴
```

**특정 열에서 작은 순서대로 불러오기**

```python
df.nsmallest(3, 'a')  # a열에서 작은 순서대로 3개를 불러옴
```
