```python
import pyupbit
import pandas as pd
import time
```

**pyupbit의 get_ohlcv를 통해 비트코인의 현재 시가를 알아오도록 하겠습니다.**

**간략히 설명해드리자면, get_ohlcv는 원하는 가상화폐의 open(시가), high(고가), low(저가), close(종가), volume(거래량)을 얻어옵니다.**

**interval을 통해 시간 단위를 설정할 수 있고, to를 통해 언제까지의 가격을 알아올지 설정할 수 있습니다.**

```python
df = pyupbit.get_ohlcv("KRW-BTC", to ="20230220")
```

**하루 단위로 불러왔는데요, 받아온 데이터를 확인해보니, 200개 정도밖에 되지 않네요**

**200개의 데이터, 즉 200일동안 이루어진 거래밖에 확인할 수 없는 것인데요. 금융데이터분석을 하기에는 너무 적은 양일 수 있죠**

**그럼 어떻게하면 더 많은 양의 데이터를 불러 올 수 있을지 알아볼까요?**

**불러온 데이터의 index를 확인해보니 시간으로 이루어져 있네요. 일단 데이터프레임의 인덱스를 확인해보죠**

```python
df.index
```

**인덱스가 DatetimeIndex로 이루어져 있네요**

**그런데 get_ohlcv의 to 파라미터에는 시간을 넣었었죠?**

**to 파라미터에 df의 첫 인덱스를 넣으면 그 다음 데이터도 불러올 수 있겠네요!**


---

**우선 첫 인덱스를 확인해보죠**

```python
index = df.index[0]
index
```

**index까지의 데이터를 볼러와볼까요?**

to 파라미터는 DateIndex의 날짜단위만 입력받습니다만, 편리하게도 따로 시간 단위를 잘라내도 되지 않습니다. (즉, to에 index를 바로 입력하면 됩니다.)


```python
pyupbit.get_ohlcv("KRW-BTC", interval="minute60", to = index)
```

**이제 for문을 이용하여 리스트에 df를 넣고 합쳐봅시다!**

```python
dfs = []
df = pyupbit.get_ohlcv("KRW-BTC", interval="minute1", to ="20230222 23:00:00")
dfs.append(df)

for i in range(5):
    df = pyupbit.get_ohlcv("KRW-BTC", to = df.index[0]) #to: 출력할 max date time을 지정
    dfs.append(df)
    time.sleep(0.1) #한 번에 너무 많은 데이터를 요청하면 웹서버에서 차단할 수 있음

df = pd.concat(dfs) #list안에 있는 데이터프레임을 하나로 합쳐줌
df = df.sort_index(ascending=False)#df를 index 기준으로 정렬
```

제대로 됐는지 확인해볼까요?
